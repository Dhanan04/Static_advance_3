Estimation statistics involves using sample data to make inferences or estimates about population parameters. Point estimate is a single value estimate of a population parameter based on sample data, such as the sample mean or sample proportion. Interval estimate provides a range of values within which the population parameter is likely to fall, along with a level of confidence associated with that range.

```python
#Ans2.)

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Assuming normal distribution and using z-distribution
    margin_of_error = 1.96 * (sample_std_dev / (sample_size ** 0.5))  # 1.96 is the z-value for 95% confidence
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound
```

#Ans3.)

Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data. It is used to determine whether there is enough evidence to reject or fail to reject a null hypothesis, which is a statement about the population parameter. Hypothesis testing helps in making decisions and drawing conclusions about populations based on sample data

#Ans4.)

Hypothesis: The average weight of male college students is greater than the average weight of female college students.

```python
#Ans5.)

import scipy.stats as stats

def hypothesis_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    if p_value < alpha:
        return True, t_stat, p_value  # Reject null hypothesis
    else:
        return False, t_stat, p_value  # Fail to reject null hypothesis
```

#Ans6.)

Null hypothesis (H0): There is no significant difference between the groups or no effect observed.

Alternative hypothesis (H1 or Ha): There is a significant difference between the groups or an effect is observed

Examples:

Null hypothesis: The mean score of Group A is equal to the mean score of Group B.
Alternative hypothesis: The mean score of Group A is not equal to the mean score of Group B.
Null hypothesis: The new drug has no effect on blood pressure.
Alternative hypothesis: The new drug has an effect on blood pressure.

#Ans7.)

Steps involved in hypothesis testing:

State the null hypothesis (H0) and alternative hypothesis (H1).

Choose the appropriate test statistic and significance level (alpha).

Collect sample data and compute the test statistic.

Determine the p-value associated with the test statistic.

Make a decision: Reject the null hypothesis if p-value < alpha, otherwise fail to reject the null hypothesis.
Draw conclusions based on the decision.

#Ans8.)

The p-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed in the sample data, assuming that the null hypothesis is true. In hypothesis testing, the p-value helps in determining the strength of evidence against the null hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis, leading to the rejection of the null hypothesis if it is below the chosen significance level (alpha).

```python
#Ans9.)

import numpy as np
import matplotlib.pyplot as plt

df = 10  # Degrees of freedom
x = np.linspace(-4, 4, 1000)  # Range of x values
y = stats.t.pdf(x, df)  # Probability density function

plt.plot(x, y)
plt.title("Student's t-distribution (df=10)")
plt.xlabel("x")
plt.ylabel("Probability density")
plt.grid(True)
plt.show()
```

```python
#Ans10.)

import scipy.stats as stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    if p_value < alpha:
        print("Reject null hypothesis: There is a significant difference between the population means.")
    else:
        print("Fail to reject null hypothesis: There is no significant difference between the population means.")

# Example usage:
sample1 = [83, 75, 84, 90, 79]
sample2 = [72, 78, 68, 82, 85]
two_sample_t_test(sample1, sample2)
```

#Ans11.)

 Student's t-distribution is a probability distribution that is used to estimate population parameters when the sample size is small or when the population standard deviation is unknown. It is similar to the standard normal distribution but has thicker tails, making it suitable for small sample sizes. The t-distribution is used in hypothesis testing and constructing confidence intervals

#Ans12.)

The t-statistic measures the difference between the sample mean and the population mean in terms of standard error.

```python
#Ans13.)

sample_mean = 500
sample_std_dev = 50
sample_size = 50
alpha = 0.05

margin_of_error = 1.96 * (sample_std_dev / (sample_size ** 0.5))  # 1.96 is the z-value for 95% confidence
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"95% Confidence Interval: (${lower_bound:.2f}, ${upper_bound:.2f})")
```

```python
#Ans14.)

sample_mean = 8
sample_std_dev = 3
sample_size = 100
hypothesized_mean = 10
alpha = 0.05

t_stat = (sample_mean - hypothesized_mean) / (sample_std_dev / (sample_size ** 0.5))
p_value = stats.t.cdf(t_stat, df=sample_size-1)  # Cumulative distribution function
if p_value < alpha:
    print("Reject null hypothesis: The new drug decreases blood pressure.")
else:
    print("Fail to reject null hypothesis: The new drug does not decrease blood pressure.")
```

```python
#Ans15.)

sample_mean = 4.8
population_mean = 5
sample_std_dev = 0.5
sample_size = 25
alpha = 0.01

t_stat = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))
p_value = stats.t.cdf(t_stat, df=sample_size-1)  # Cumulative distribution function
if p_value < alpha:
    print("Reject null hypothesis: The true mean weight of the products is less than 5 pounds.")
else:
    print("Fail to reject null hypothesis: The true mean weight of the products is greater than or equal to 5 pounds.")
```

```python
#Ans16.)
mean1 = 80
std_dev1 = 10
n1 = 30

mean2 = 75
std_dev2 = 8
n2 = 40

alpha = 0.01

t_stat = (mean1 - mean2) / ((std_dev1**2/n1 + std_dev2**2/n2)**0.5)
p_value = stats.t.cdf(t_stat, df=n1+n2-2)  # Cumulative distribution function
if p_value < alpha:
    print("Reject null hypothesis: The population means for the two groups are not equal.")
else:
    print("Fail to reject null hypothesis: The population means for the two groups are equal.")
```
